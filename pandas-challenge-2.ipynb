```python
import pandas as pd
from pathlib import Path
```

```python
schooldata = Path("Resources/schools_complete.csv")
studentdata = Path("Resources/students_complete.csv")

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(schooldata)
student_data = pd.read_csv(studentdata)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

```python
school_count= pd.DataFrame(school_data['school_name'])
school_count1=len(school_count.index)
school_count1
```

```python
student_count=pd.DataFrame(student_data['student_name'])
student_count1=len(student_count.index)
student_count1
```

```python
total_budget=school_data['budget']
total_budget1=sum(total_budget)
total_budget1
```

```python
math_average=student_data['math_score']
math_average1=math_average.mean()
math_average1
```

```python
score_reading=student_data['reading_score']
score_reading1=score_reading.mean()
score_reading1
```

```python
passing_math_count=student_data[(student_data["math_score"] >= 70)].count()['student_name']
passing_math_count1=(passing_math_count/len(student_count.index))*100
passing_math_count1
```

```python
passing_reading_count=student_data[(student_data["reading_score"] >= 70)].count()['student_name']
passing_reading_count1=(passing_reading_count/len(student_count.index))*100
passing_reading_count1
```

```python
passing_math_reading_count = student_data[
    (student_data["math_score"] >= 70) & (student_data["reading_score"] >= 70)
].count()["student_name"]
overall_passing_rate = passing_math_reading_count /  len(student_count.index)* 100
overall_passing_rate
```

```python
columns=['total school','total students','total budget','average math score','average reading score','%passing math','% passing reading','%overall passing']
data=[school_count1,student_count1,total_budget1,math_average1,score_reading1,passing_math_count1,passing_reading_count1,overall_passing_rate]

district_summary = pd.DataFrame(data)
district_summary=district_summary.transpose()
district_summary.columns=columns


#district_summary["total students"] = district_summary["total students"].map("{:,}".format)
#district_summary["total budget"] = district_summary["total budget"].map("${:,.2f}".format)


district_summary
```

# SCHOOL SUMMARY

```python
#school_type=school_data_complete['type']
school_type1=pd.DataFrame(school_data_complete[['school_name','type']]).drop_duplicates()
school_type1
```

```python
per_school_count=pd.DataFrame(student_data['school_name'])
per_school_count1=per_school_count.value_counts()
per_school_count1=pd.DataFrame(per_school_count1)
per_school_count1.columns=['#students']
per_school_count1
```

```python
per_school_budget=pd.DataFrame(school_data_complete[['school_name','budget']])
per_school_budget1=pd.DataFrame(per_school_budget).drop_duplicates()
per_school_budget1 = per_school_budget1.set_index('school_name')
per_school_budget1

```

```python
per_capita=pd.merge(per_school_count1,per_school_budget1, how="left", on=[ "school_name"])
per_capita1=per_capita['budget']/per_capita['#students']
per_capita1=pd.DataFrame(per_capita1)

per_capita1.columns=['Per Student Budget']
per_capita1
```

```python
per_school_math=pd.DataFrame(student_data[['school_name','math_score']])
per_school_math1=per_school_math.groupby(['school_name']).mean()
per_school_math1
```

```python
per_school_reading=pd.DataFrame(student_data[['school_name','reading_score']])
per_school_reading1=per_school_reading.groupby(['school_name']).mean()
per_school_reading1
```

```python
students_passing_math=pd.DataFrame(school_data_complete[(school_data_complete["math_score"] >= 70)])
students_passing_math=students_passing_math[['school_name','math_score']]
students_passing_math1=students_passing_math.groupby(['school_name']).count()
students_passing_math2=pd.merge(students_passing_math1,per_school_count1, how="left", on=["school_name"])
students_passing_math3=students_passing_math2['math_score']/students_passing_math2['#students']*100
students_passing_math3=pd.DataFrame(students_passing_math3)
students_passing_math3.columns = ['% Passing math']
students_passing_math1
```

```python
students_passing_reading=pd.DataFrame(school_data_complete[(school_data_complete["reading_score"] >= 70)])
students_passing_reading=students_passing_reading[['school_name','reading_score']]
students_passing_reading1=students_passing_reading.groupby(['school_name']).count()
students_passing_reading2=pd.merge(students_passing_reading1,per_school_count1, how="left", on=["school_name"])
students_passing_reading3=students_passing_reading2['reading_score']/students_passing_reading2['#students']*100
students_passing_reading3=pd.DataFrame(students_passing_reading3)
students_passing_reading3.columns = ['% Passing reading']
students_passing_reading1
```

```python
students_passing_math3
students_passing_reading3 = pd.DataFrame(students_passing_reading3)
students_passing_reading3.columns = ['% Passing reading']
students_passing_reading3
```

```python
students_passing_math_and_reading = school_data_complete[
    (school_data_complete["reading_score"] >= 70) & (school_data_complete["math_score"] >= 70)
]
school_students_passing_math_and_reading = pd.DataFrame(students_passing_math_and_reading.groupby(["school_name"]).size())
school_students_passing_math_and_reading.columns = ['count']

school_students_passing_math_and_reading2=pd.merge(school_students_passing_math_and_reading,per_school_count1, how="left", on=["school_name"])
school_students_passing_math_and_reading3=school_students_passing_math_and_reading2['count']/school_students_passing_math_and_reading2['#students']*100
school_students_passing_math_and_reading3=pd.DataFrame(school_students_passing_math_and_reading3)
school_students_passing_math_and_reading3.columns = ['% Overall Passing']
school_students_passing_math_and_reading3
```

```python
per_school_summary = pd.merge(school_type1, per_school_count1, how="left", on=["school_name"]).merge(per_school_budget1,how="left", on=["school_name"]).merge(per_capita1,how="left", on=["school_name"]).merge(per_school_math1,how="left", on=["school_name"]).merge(per_school_reading1,how="left", on=["school_name"]).merge(students_passing_math3,how="left", on=["school_name"]).merge(students_passing_reading3,how="left", on=["school_name"]).merge(school_students_passing_math_and_reading3,how="left", on=["school_name"])
per_school_summary["budget"] = per_school_summary["budget"].map("${:,.2f}".format)
per_school_summary["Per Student Budget"] = per_school_summary["Per Student Budget"].map("${:,.2f}".format)

per_school_summary
```

```python
top_schools = per_school_summary.sort_values('% Overall Passing',ascending=False)
top_schools.head(5)
```

```python
bottom_schools = per_school_summary.sort_values('% Overall Passing',ascending=True)
bottom_schools.head(5)
```
